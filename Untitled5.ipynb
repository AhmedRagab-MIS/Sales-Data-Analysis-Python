```python
df['Discount'] = df['Discount'].fillna(0)
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['UnitPrice'] = df['UnitPrice'].fillna(df['UnitPrice'].median())
```

```python
df.isnull().sum()
```

```python
df = df[df['Quantity'] > 0]
df
```

```python
df.describe()
```

```python
df['Total_Sales'] = df['Quantity'] * df['UnitPrice'] * (1 - df['Discount'])
```

```python
df
```

```python
df.groupby('Region')['Total_Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Category')['Total_Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Customer')['Total_Sales'].sum().sort_values(ascending=False).head(5)
```

```python
pd.set_option('display.float_format' , lambda x: '%.2f' % x)
```

```python
import matplotlib.pyplot as plt
category_Sales = df.groupby('Category')['Total_Sales'].sum()
category_Sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.ylabel('')
plt.show()
```
